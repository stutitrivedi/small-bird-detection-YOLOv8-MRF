```python
frame1 = cv2.imread('/content/drive/My Drive/Thesis/Abs_Difference/Image1.JPG/',cv2.IMREAD_GRAYSCALE)

frame2 = cv2.imread('/content/drive/My Drive/Thesis/Abs_Difference/Image2.JPG', cv2.IMREAD_GRAYSCALE)
!ls '/content/drive/My Drive/Thesis/Abs_Difference/Image1.JPG'

import cv2

# Load the images
frame1 = cv2.imread('/content/drive/My Drive/Thesis/Abs_Difference/Image1.JPG', cv2.IMREAD_GRAYSCALE)
frame2 = cv2.imread('/content/drive/My Drive/Thesis/Abs_Difference/Image2.JPG', cv2.IMREAD_GRAYSCALE)

# Check the dimensions of the images
print("Frame 1 shape:", frame1.shape)
print("Frame 2 shape:", frame2.shape)

# Ensure the images have the same dimensions
if frame1.shape == frame2.shape:
    # Compute the pixel changes
    delta_frame = cv2.absdiff(frame2, frame1)
    # Continue with the rest of your code...
else:
    print("Error: Images have different dimensions.")

# Compute the pixel changes
#delta_frame = cv2.absdiff(frame2, frame1)

output_folder = '/content/drive/My Drive/Thesis/Abs_Difference/output_folder'
os.makedirs(output_folder, exist_ok=True)
# List of threshold values
threshold_values = [5, 10, 20, 30, 40]

for thresh_val in threshold_values:
    # Threshold delta_frame
    thresh_delta = cv2.threshold(delta_frame, thresh_val, 255, cv2.THRESH_BINARY)[1]

    # Find contours in the thresholded image
    contours, _ = cv2.findContours(thresh_delta, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Analyze the pixel cluster
    for contour in contours:
        if cv2.contourArea(contour) < 100:
            continue
        (x, y, w, h) = cv2.boundingRect(contour)
        cv2.rectangle(frame1, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Display and save the images
print("Threshold Value:", thresh_val)
cv2_imshow(frame1)
cv2.imwrite(os.path.join(output_folder, f'frame1_thresh_{thresh_val}.png'), frame1)
cv2_imshow(frame2)
cv2.imwrite(os.path.join(output_folder, f'frame2_thresh_{thresh_val}.png'), frame2)
cv2_imshow(delta_frame)
cv2.imwrite(os.path.join(output_folder, f'delta_frame_thresh_{thresh_val}.png'), delta_frame)
cv2_imshow(thresh_delta)
cv2.imwrite(os.path.join(output_folder, f'thresh_delta_{thresh_val}.png'), thresh_delta)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
